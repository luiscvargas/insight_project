```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from api_import import *
```

```python
df = pd.read_json("../data/census_data.json")
df.set_index(['COUNTY_ID', 'TRACT_ID'],inplace=True,drop=True)
df.head(5)
```

```python
ziplist = df.ZIPCODE.astype(str).unique()
ziplist
```

```python
zillow_data = query_zillow(ziplist)
```

```python
zillow_data.items()
```

```python
df['zillow_house_median'] = 0.0
df['zillow_house_index'] = 0.0
```

```python
zillow_zips = zillow_data.keys()
for irow, row in df.iterrows():
    if str(row['ZIPCODE']) in zillow_zips:
        df.loc[irow,'house_index'] = zillow_data[str(row['ZIPCODE'])][0]
        df.loc[irow,'house_median'] = zillow_data[str(row['ZIPCODE'])][1]
    else:
        df.loc[irow,'house_index'] = np.nan
        df.loc[irow,'house_median'] = np.nan
```
